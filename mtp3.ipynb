```python
min_lat=df['latitude [dec.deg.N]'].min()
max_lat=df['latitude [dec.deg.N]'].max()
min_lon=df['longitude [dec.deg.E]'].min()
max_lon=df['longitude [dec.deg.E]'].max()
print(f'min latitude:{min_lat},max latitude:{max_lat}')
print(f'min longitude:{min_lon},max longitude:{max_lon}')
```

```python
df=df.rename(columns={'latitude [dec.deg.N]':'lat','longitude [dec.deg.E]':'lon','SST [deg.C]':'SST','sal':'Salinity','yr':'year','day':'day','mon':'month','hh':'hour','mm':'minute','ss':'second','pCO2water_SST_wet [uatm]':'pCO2'})
df
```

```python
df.info()
```

```python
features=df[['lat','lon','SST','Salinity','year','day','month','hour','minute','second','pCO2']]
features
```

```python
features['pCO2'].fillna(features['pCO2'].median(),inplace=True)
```

```python
features['Salinity'].fillna(features['Salinity'].median(),inplace=True)
```

```python
features['SST'].fillna(features['SST'].mean(),inplace=True)
```

```python
features['time'] = pd.to_datetime(features[['year', 'month', 'day', 'hour', 'minute', 'second']])

features.drop(columns=['year', 'month', 'day', 'hour', 'minute', 'second'], inplace=True)
```

```python
arabian_sea_lat_range = (-1, 26)
arabian_sea_lon_range = (51, 75)

bay_of_bengal_lat_range = (5, 25)
bay_of_bengal_lon_range = (78, 96)
```

```python
# Filter data for Arabian Sea
df_arabian_sea = features[(features['lat'] >= arabian_sea_lat_range[0]) & (features['lat'] <= arabian_sea_lat_range[1]) &
                    (features['lon'] >= arabian_sea_lon_range[0]) & (features['lon'] <= arabian_sea_lon_range[1])]
```

```python
# Filter data for Bay of Bengal
df_bay_of_bengal = features[(features['lat'] >= bay_of_bengal_lat_range[0]) & (features['lat'] <= bay_of_bengal_lat_range[1]) &
                      (features['lon'] >= bay_of_bengal_lon_range[0]) & (features['lon'] <= bay_of_bengal_lon_range[1])]
```

```python
arabian_sea_corr = df_arabian_sea[['SST', 'Salinity', 'pCO2']].corr()
arabian_sea_corr
```

```python
bay_of_bengal_corr = df_bay_of_bengal[['SST', 'Salinity', 'pCO2']].corr()
bay_of_bengal_corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))

# Plot for the Arabian Sea
plt.subplot(1, 2, 1)
sns.heatmap(arabian_sea_corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix - Arabian Sea')

# Plot for the Bay of Bengal
plt.subplot(1, 2, 2)
sns.heatmap(bay_of_bengal_corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix - Bay of Bengal')

# Display the plots
plt.tight_layout()
plt.savefig('heatmap.png', dpi=300)
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import cartopy.crs as ccrs
import cartopy.feature as cfeature


# Convert DataFrame to GeoDataFrame with correct CRS
features['geometry'] = features.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
gfeatures = gpd.GeoDataFrame(features, geometry='geometry', crs="EPSG:4326")

# Define the boundaries for the Arabian Sea and Bay of Bengal using conditional filtering
gdf_arabian_sea = gfeatures[(gfeatures['lon'] >= 51) & (gfeatures['lon'] <= 75) &
                            (gfeatures['lat'] >= -1) & (gfeatures['lat'] <= 26)]
gdf_bay_of_bengal = gfeatures[(gfeatures['lon'] >= 78) & (gfeatures['lon'] <= 96) &
                              (gfeatures['lat'] >= 5) & (gfeatures['lat'] <= 25)]

# Find common vmin and vmax for Salinity and SST to ensure consistent color scales
min_salinity = min(gdf_arabian_sea['Salinity'].min(), gdf_bay_of_bengal['Salinity'].min())
max_salinity = max(gdf_arabian_sea['Salinity'].max(), gdf_bay_of_bengal['Salinity'].max())

min_sst = min(gdf_arabian_sea['SST'].min(), gdf_bay_of_bengal['SST'].min())
max_sst = max(gdf_arabian_sea['SST'].max(), gdf_bay_of_bengal['SST'].max())

min_pco2 = min(gdf_arabian_sea['pCO2'].min(), gdf_bay_of_bengal['pCO2'].min())
max_pco2 = max(gdf_arabian_sea['pCO2'].max(), gdf_bay_of_bengal['pCO2'].max())

# Create a figure for plotting
fig, axes = plt.subplots(3, 2, figsize=(15, 12), subplot_kw={'projection': ccrs.PlateCarree()})

extent1 = [50, 76,-2, 26]
extent2 = [78, 96,-2, 26]

axes[0, 0].set_extent(extent1)

# Plot for Salinity - Arabian Sea
gdf_arabian_sea.plot(column='Salinity', ax=axes[0, 0], legend=True,
                     cmap='viridis',vmin=min_salinity,vmax=max_salinity,legend_kwds={'label': "Salinity (ppt)"})
axes[0, 0].set_title('Arabian Sea - Salinity')
axes[0, 0].add_feature(cfeature.COASTLINE)
axes[0, 0].add_feature(cfeature.BORDERS, linestyle=':')
axes[0, 0].gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

axes[0, 1].set_extent(extent2)

# Plot for Salinity - Bay of Bengal
gdf_bay_of_bengal.plot(column='Salinity', ax=axes[0, 1], legend=True,
                       cmap='viridis',vmin=min_salinity,vmax=max_salinity,legend_kwds={'label': "Salinity (ppt)"})
axes[0, 1].set_title('Bay of Bengal - Salinity')
axes[0, 1].add_feature(cfeature.COASTLINE)
axes[0, 1].add_feature(cfeature.BORDERS, linestyle=':')
axes[0, 1].gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

axes[1, 0].set_extent(extent1)

# Plot for SST - Arabian Sea
sst_arabian_sea = gdf_arabian_sea.plot(column='SST', ax=axes[1, 0], legend=True,
                                        cmap='coolwarm',vmin=min_sst,vmax=max_sst,legend_kwds={'label': "SST (°C)"})
axes[1, 0].set_title('Arabian Sea - SST')
axes[1, 0].add_feature(cfeature.COASTLINE)
axes[1, 0].add_feature(cfeature.BORDERS, linestyle=':')
axes[1, 0].gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

axes[1, 1].set_extent(extent2)
# Plot for SST - Bay of Bengal
sst_bay_of_bengal = gdf_bay_of_bengal.plot(column='SST', ax=axes[1, 1], legend=True,
                                           cmap='coolwarm',vmin=min_sst,vmax=max_sst,legend_kwds={'label': "SST (°C)"})
axes[1, 1].set_title('Bay of Bengal - SST')
axes[1, 1].add_feature(cfeature.COASTLINE)
axes[1, 1].add_feature(cfeature.BORDERS, linestyle=':')
axes[1, 1].gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

axes[2, 0].set_extent(extent1)

# Plot for pCO2 - Arabian Sea
gdf_arabian_sea.plot(column='pCO2', ax=axes[2, 0], legend=True,
                     cmap='viridis',vmin=min_pco2,vmax=max_pco2,legend_kwds={'label': "pCO2 (uatm)"})
axes[2, 0].set_title('Arabian Sea - pCO2')
axes[2, 0].add_feature(cfeature.COASTLINE)
axes[2, 0].add_feature(cfeature.BORDERS, linestyle=':')
axes[2, 0].gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

axes[2, 1].set_extent(extent2)

# Plot for Salinity - Bay of Bengal
gdf_bay_of_bengal.plot(column='pCO2', ax=axes[2, 1], legend=True,
                       cmap='viridis',vmin=min_pco2,vmax=max_pco2,legend_kwds={'label': "pCO2 (uatm)"})
axes[2, 1].set_title('Bay of Bengal - pCO2')
axes[2, 1].add_feature(cfeature.COASTLINE)
axes[2, 1].add_feature(cfeature.BORDERS, linestyle=':')
axes[2, 1].gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)


plt.subplots_adjust(right=1.9)
plt.subplots_adjust(wspace=0.9)
#plt.subplots_adjust(hspace=1)
# Adjust layout and show plot

plt.tight_layout()
plt.savefig('output1.png', dpi=300)
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from shapely.geometry import Point

# Convert DataFrame to GeoDataFrame
features['geometry'] = features.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
gdf = gpd.GeoDataFrame(features, geometry='geometry')

# Define boundaries for Arabian Sea and Bay of Bengal
gdf_arabian_sea = gdf.cx[51:75, -1:26]
gdf_bay_of_bengal = gdf.cx[78:96, 5:25]

# Find common vmin and vmax for Salinity and SST to ensure consistent color scales
min_salinity = min(gdf_arabian_sea['Salinity'].min(), gdf_bay_of_bengal['Salinity'].min())
max_salinity = max(gdf_arabian_sea['Salinity'].max(), gdf_bay_of_bengal['Salinity'].max())

min_sst = min(gdf_arabian_sea['SST'].min(), gdf_bay_of_bengal['SST'].min())
max_sst = max(gdf_arabian_sea['SST'].max(), gdf_bay_of_bengal['SST'].max())

min_pco2 = min(gdf_arabian_sea['pCO2'].min(), gdf_bay_of_bengal['pCO2'].min())
max_pco2 = max(gdf_arabian_sea['pCO2'].max(), gdf_bay_of_bengal['pCO2'].max())

# Create a plot with Cartopy projections
fig, axes = plt.subplots(3, 2, figsize=(17, 15), subplot_kw={'projection': ccrs.PlateCarree()})

# Define common plotting options
extent = [50, 96, -2, 26]  # Extent covering both regions

# Plot Salinity for Arabian Sea (Top Left)
axes[0, 0].set_extent(extent)
axes[0, 0].add_feature(cfeature.COASTLINE)
sc1 = axes[0, 0].scatter(gdf_arabian_sea['lon'], gdf_arabian_sea['lat'], c=gdf_arabian_sea['Salinity'],
                         cmap='viridis', vmin=min_salinity,vmax=max_salinity)
axes[0, 0].set_title("Arabian Sea - Salinity")
plt.colorbar(sc1, ax=axes[0, 0], orientation='horizontal',label="Salinity (ppt)", shrink=0.54)

# Plot Salinity for Bay of Bengal (Top Right)
axes[0, 1].set_extent(extent)
axes[0, 1].add_feature(cfeature.COASTLINE)
sc2 = axes[0, 1].scatter(gdf_bay_of_bengal['lon'], gdf_bay_of_bengal['lat'], c=gdf_bay_of_bengal['Salinity'],
                         cmap='viridis', vmin=min_salinity,vmax=max_salinity)
axes[0, 1].set_title("Bay of Bengal - Salinity")
plt.colorbar(sc2, ax=axes[0, 1], orientation='horizontal',label="Salinity (ppt)", shrink=0.54)

# Plot SST for Arabian Sea (middle Left)
axes[1, 0].set_extent(extent)
axes[1, 0].add_feature(cfeature.COASTLINE)
sc3 = axes[1, 0].scatter(gdf_arabian_sea['lon'], gdf_arabian_sea['lat'], c=gdf_arabian_sea['SST'],
                         cmap='coolwarm',vmin=min_sst,vmax=max_sst)
axes[1, 0].set_title("Arabian Sea - SST")
plt.colorbar(sc3, ax=axes[1, 0], orientation='horizontal',label="SST (°C)", shrink=0.54)

# Plot SST for Bay of Bengal (middle Right)
axes[1, 1].set_extent(extent)
axes[1, 1].add_feature(cfeature.COASTLINE)
sc4 = axes[1, 1].scatter(gdf_bay_of_bengal['lon'], gdf_bay_of_bengal['lat'], c=gdf_bay_of_bengal['SST'],
                         cmap='coolwarm',vmin=min_sst,vmax=max_sst)
axes[1, 1].set_title("Bay of Bengal - SST")
plt.colorbar(sc4, ax=axes[1, 1], orientation='horizontal',label="SST (°C)", shrink=0.54)

# Plot pCO2 for Arabian Sea (bottom Left)
axes[2, 0].set_extent(extent)
axes[2, 0].add_feature(cfeature.COASTLINE)
sc5 = axes[2, 0].scatter(gdf_arabian_sea['lon'], gdf_arabian_sea['lat'], c=gdf_arabian_sea['pCO2'],
                         cmap='viridis', vmin=min_pco2,vmax=max_pco2)
axes[2, 0].set_title("Arabian Sea - pCO2")
plt.colorbar(sc5, ax=axes[2, 0], orientation='horizontal',label="pCO2 (uatm)", shrink=0.54)

# Plot pCO2 for Bay of Bengal (bottom Right)
axes[2, 1].set_extent(extent)
axes[2, 1].add_feature(cfeature.COASTLINE)
sc6 = axes[2, 1].scatter(gdf_bay_of_bengal['lon'], gdf_bay_of_bengal['lat'], c=gdf_bay_of_bengal['pCO2'],
                         cmap='viridis', vmin=min_pco2,vmax=max_pco2)
axes[2, 1].set_title("Bay of Bengal - pCO2")
plt.colorbar(sc6, ax=axes[2, 1], orientation='horizontal',label="pCO2 (uatm)", shrink=0.54)

# Add additional map features
for ax in axes.flat:
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', linestyle='--')

# Layout adjustments
plt.tight_layout()
plt.savefig('output2.png', dpi=300)
plt.show()
```
