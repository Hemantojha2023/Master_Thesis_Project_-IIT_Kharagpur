__Let's Impliment XGBoost on Bay of Bengal for pCO2__

```python
import pandas as pd
df1=pd.read_csv('target_pCO2.csv')
```

```python
df1
```

```python
df1=df1.drop('Unnamed: 0', axis=1)
```

```python
df1.isna().sum()
```

```python
x=df1.drop('pCO2_SST', axis=1)
y=df1['pCO2_SST']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
xgb_model=XGBRegressor(n_estimators=600, max_depth=4, learning_rate=0.02, eval_metric='rmse', random_state=42)
xgb_model.fit(x_train, y_train)
```

```python
import numpy as np
y_train_pred=xgb_model.predict(x_train)
y_test_pred=xgb_model.predict(x_test)

training_accuracy=np.sqrt(mean_squared_error(y_train, y_train_pred))
testing_accuracy=np.sqrt(mean_squared_error(y_test, y_test_pred))

print('XGBoost Regression :')
print('Training Error :',training_accuracy)
print('Testing Error :',testing_accuracy)
```

```python
from sklearn.model_selection import KFold, cross_val_score
kfold=KFold(n_splits=10, shuffle=True, random_state=42)
cv_scores=cross_val_score(xgb_model, x_train, y_train, cv=kfold)
print('Cross Validation Scores :',cv_scores)
print('CV Scores Mean :',cv_scores.mean())
```

Model is more consistent and low variance

Plot actual vs predicted data points with best fit line 

- If the model is perfect, points will lie on a 45-degree diagonal line.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.regplot(x=y_test, y=y_test_pred, scatter_kws={"color": "blue", "alpha": 0.6}, line_kws={"color": "red"})
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Least Squares Approximation (Actual vs. Predicted)")
plt.show()
```

We can see that the most point lies on the regressoin line so this line is the best fit line and model best fit model

__Let's impliment XGBRegressor on the data which has 15% nan value__

```python
x=new_df.drop('pCO2_SST', axis=1)
y=new_df['pCO2_SST']
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
x['QC_Flag']=encoder.fit_transform(x['QC_Flag'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

XGBoost handles missing values in features (X) automatically. But, the target variable (y) cannot contain NaN values. XGBoost does not accept NaN values in y_train

According to standard machine learning practices, filling missing values in the test set with statistics (like median) calculated from the training set does not constitute Data leakage. This method ensures that no information from the test set influences the training process, thus maintaining the integrity of the evaluation.

```python
median_pCO2=y_train.median(skipna=True)
y_train=y_train.fillna(median_pCO2)
y_test=y_test.fillna(median_pCO2)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
xgb_model=XGBRegressor(n_estimators=600, max_depth=4, learning_rate=0.02, eval_metric='rmse', random_state=42)
xgb_model.fit(x_train, y_train)
```

```python
y_train_pred=xgb_model.predict(x_train)
y_test_pred=xgb_model.predict(x_test)

training_accuracy=np.sqrt(mean_squared_error(y_train, y_train_pred))
testing_accuracy=np.sqrt(mean_squared_error(y_test, y_test_pred))

print('XGBoost Regression :')
print('Training Error :',training_accuracy)
print('Testing Error :',testing_accuracy)
```

model is generalize well on training set but not on testing set so model overfitted 

```python
from sklearn.model_selection import KFold, cross_val_score
kfold=KFold(n_splits=10, shuffle=True, random_state=42)
cv_scores=cross_val_score(xgb_model, x_train, y_train, cv=kfold)
print('Cross Validation Scores :',cv_scores)
print('CV Scores Mean :',cv_scores.mean())
```

CV scores is consistent so model is consistent and low variance

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.regplot(x=y_test, y=y_test_pred, scatter_kws={"color": "blue", "alpha": 0.6}, line_kws={"color": "red"})
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Least Squares Approximation (Actual vs. Predicted)")
plt.show()
```
